```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn')

data = {'one': [30, 20, 10, 15, 25],
        'two': [40, 10, 20, 15, 15],
        'three': [15, 15, 10, 40, 20],
        'four': [15, 20, 30, 25, 10]}

df = pd.DataFrame(data, index=list('ABCDE'))

exp = [0, 0.2, 0, 0, 0.3]

c = ['lightgreen', 'cyan', 'pink', 'lightblue', 'wheat']

plt.subplot(2, 2, 1)
df['one'].plot.pie(fontsize=14, autopct='%1.1f%%', explode=exp, 
                   colors=c)

plt.subplot(2, 2, 2)
df['two'].plot.pie(fontsize=14, autopct='%1.1f%%', explode=exp, 
                   colors=c)

plt.subplot(2, 2, 3)
df['three'].plot.pie(fontsize=14, autopct='%1.1f%%', explode=exp, 
                     colors=c)

plt.subplot(2, 2, 4)
df['four'].plot.pie(fontsize=14, autopct='%1.1f%%', explode=exp, 
                    colors=c)

plt.subplots_adjust(wspace=0.5, hspace=0.4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn')

data = {'group': ['B', 'C', 'B', 'A', 'A', 'C', 'B', 'C', 'A'],
        'name': ['Jim', 'John', 'Jack', 'Jane', 'Joe', 'Jake', 
                 'Jeff', 'Jill', 'James'],
        'score': [70, 60, 80, 50, 70, 50, 90, 70, 80]}

df = pd.DataFrame(data)

df.groupby('group').sum().plot(kind='bar', rot=0, color='c')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn')

data = {'group': ['B', 'C', 'B', 'A', 'A', 'C', 'B', 'C', 'A', 'B'],
        'gender': ['Male', 'Male', 'Male', 'Female', 'Male', 
                   'Female', 'Male', 'Female', 'Male', 'Female'],
        'name': ['Jim', 'John', 'Jack', 'Jane', 'Joe', 'Joy', 
                 'Jeff', 'Jill', 'James', 'Jan'],
        'score': [70, 60, 80, 50, 70, 50, 90, 70, 80, 90]}

df = pd.DataFrame(data)
#display(df.sort_values('gender'))

pv = df.pivot_table(index='group',
                    columns='gender', 
                    values='score',
                    aggfunc='sum').add_suffix('_sum')

display(pv)

pv.plot(style=['ro:', 'go--'])
```
